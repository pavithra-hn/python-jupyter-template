### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.

import matplotlib.pyplot as plt
import numpy as np

try:
    # Generate x values from 0 to 10
    x = np.linspace(0, 10, 100)
    # Calculate y values as the square of x
    y = x ** 2

    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='y = x^2', color='blue')

    # Customize the plot
    plt.title('Line Plot of y = x^2', fontsize=16)
    plt.xlabel('X values', fontsize=14)
    plt.ylabel('Y values (Square of X)', fontsize=14)
    plt.grid(True)
    plt.legend()

    # Show the plot
    plt.show()

except ImportError as e:
    print("Error: Required library not found. Please ensure matplotlib and numpy are installed.")
except Exception as e:
    print(f"An error occurred: {e}")
```
