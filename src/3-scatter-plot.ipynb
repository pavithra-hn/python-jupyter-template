### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.

import matplotlib.pyplot as plt

try:
    # Data
    x = [1, 2, 3, 4, 5]
    y = [2, 4, 6, 8, 10]

    # Create the scatter plot
    plt.figure(figsize=(8, 5))
    plt.scatter(x, y, color='blue', s=100)  # s is the size of the points

    # Annotate each point with its (x, y) value
    for i in range(len(x)):
        plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", 
                     xytext=(0, 10), ha='center')

    # Add title and labels
    plt.title('Scatter Plot with Annotations', fontsize=16)
    plt.xlabel('X values', fontsize=14)
    plt.ylabel('Y values', fontsize=14)

    # Show the plot
    plt.grid(True)
    plt.show()

except ImportError as e:
    print("Error: Required library not found. Please ensure matplotlib is installed.")
except Exception as e:
    print(f"An error occurred: {e}")
```
