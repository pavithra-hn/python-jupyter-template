### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt

try:
    # Data
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]

    # Create the bar plot
    plt.figure(figsize=(8, 5))
    bars = plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])

    # Add title and labels
    plt.title('Bar Plot', fontsize=16)
    plt.xlabel('Categories', fontsize=14)
    plt.ylabel('Values', fontsize=14)

    # Show the plot
    plt.show()

except ImportError as e:
    print("Error: Required library not found. Please ensure matplotlib is installed.")
except Exception as e:
    print(f"An error occurred: {e}")
```
